```python
#constructing dataset
from transformers import LongformerConfig, LongformerForSequenceClassification,LongformerTokenizer
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import numpy as np
import jsonlines as js
tokenizer = LongformerTokenizer.from_pretrained("allenai/longformer-base-4096")
class text_data(Dataset):
    def __init__(self,path):
        self.file_path = path
        f=open(path,"r")
        self.json=[x for x in js.Reader(f)]
        f.close()
        self.labels=[line['label'] for line in self.json]
        self.texts= [tokenizer(line['text'], max_length = 4096, truncation=True,return_tensors="pt") for line in self.json]
    def __getitem__(self,index):
        text=self.texts[index]
        label=[0,0]
        label[self.labels[index]]=1
        return text,torch.FloatTensor(label)

    def __len__(self):
        return len(self.json)
train_set=text_data('../data/train.jsonl')
train_loader=DataLoader(train_set, batch_size=1)
valid_set=text_data('../data/fake_news_validation.jsonl')
valid_loader=DataLoader(valid_set, batch_size=1)
test_set=text_data('../data/fake_news_test.jsonl')
test_loader=DataLoader(test_set, batch_size=1)
```

```python
#LongFormer
from torch import optim as opt
import torch.nn as nn
import torch.nn.functional as F
config=LongformerConfig.from_pretrained("allenai/longformer-base-4096")
tokenizer = LongformerTokenizer.from_pretrained("allenai/longformer-base-4096",truncation=True,padding='max_length', max_length=4096)
model = LongformerForSequenceClassification(config).cuda()
optimizer=opt.Adam(model.parameters(),lr=0.001)
Loss_f=nn.CrossEntropyLoss()
epochs=20
train_acc=[]
valid_acc=[]
for i in range(0,epochs):
    x=0
    correct_case=0
    model.train()
    for text,label in train_loader:
        if x>1000:
            break
        input =text['input_ids'].squeeze(1).cuda()
        optimizer.zero_grad()
        pred=(model(input).logits).cuda()
        label=label.cuda()
        loss=Loss_f(pred,label).cuda()
        loss.backward()
        optimizer.step()
        x+=1
        if (pred.argmax()==label.argmax()):
            correct_case+=1
    print (correct_case)
    train_acc.append(correct_case/100)
    correct_case=0
    model.eval()
    for text,label in valid_loader:
        input =text['input_ids'].squeeze(1).cuda()
        pred=(model(input).logits)
        if (pred.argmax()==label.argmax()):
            correct_case+=1
    valid_acc.append(correct_case/len(valid_set))
'''
    for text,label in train_loader:
        input = tokenizer(text, return_tensors="pt")
        pred=F.softmax(model(input).logits).squeeze(0)
        loss=Loss_f(pred,label)
        print(loss)
        loss.backward()
        optimizer.step()
'''
'''
with torch.no_grad():
    logits = model(**inputs).logits
predicted_class_id = logits.argmax().item()
model.config.id2label[predicted_class_id]
'''
```
